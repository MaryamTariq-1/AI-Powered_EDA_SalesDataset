```python
# Step 1: Setup, load and quick inventory
# helpful for displaying outputs except print()
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
# Display settings
sns.set(style="whitegrid", font_scale=1.05)
# sns.set() configure seaborn/matplotlib styling for plot look clean grid and font_scale=1.05 --? larger text
pd.options.display.max_columns = None # print all columns
pd.options.display.width = 160 # text width

# Load dataset (adjust path if needed)
# encoding="latin1" specifies file encoding to avoid unicodeDecodeError
df = pd.read_csv("../data/sample_dataset.csv", encoding="latin1")

# Basic checks
#gives dataset size
print("Shape:", df.shape)
# show column name as python list
print("\nColumns:\n", df.columns.tolist())

print("\n--- HEAD (5) ---")
display(df.head())

print("\n--- RANDOM SAMPLE (5) ---")
display(df.sample(5, random_state=42)) # random_stae=42 ensure same sample every run

print("\n--- INFO ---")
# will print index dtype, column anmes, number of non-null entries per column, data type of each column, and memory storage
df.info()

#checking missing values
# a small table listing columns that have any missing values and how many/what percent. This directs cleanup priorities.
print("\n--- Missing values (count & %) ---")
missing = df.isnull().sum()
missing_pct = (df.isnull().mean()*100).round(2)
missing_df = pd.concat([missing, missing_pct], axis=1)
missing_df.columns = ['missing_count','missing_pct']
display(missing_df[missing_df['missing_count']>0].sort_values('missing_count', ascending=False))


# Create a missing data matrix plot
print("Generating Missing Data Matrix...")
msno.matrix(df)

# Save the plot to the specified directory
output_path = r"C:\Users\HP 840G4\PycharmProjects\EDA\Visualization\missing_data_matrix.png"
print(f"Saving missing data matrix plot to {output_path}...")
plt.savefig(output_path)

# Display the plot
plt.show()

print("Missing data matrix plot saved successfully!")


# returns unique values
print("\n--- Number of unique values per column ---")
display(df.nunique().sort_values(ascending=False).to_frame(name='n_unique'))

# gives count, mean, std, min, 25%, 50%, 75%,max
print("\n--- Numeric summary (describe) ---")
display(df.describe())
print("\n--- Duplicate Rows ---")
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Dataset has 2823 rows, and 25 columns.
# found 4 columns with missing values in ADDRESSLINE2, STATE, TERRITORY, POSTALCODE

# Step 2: Data Cleaning & Preparation

# Convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')

# Drop ADDRESSLINE2 (too many missing values)
df = df.drop(columns=['ADDRESSLINE2'])

# Quick check for ORDERDATE conversion
print("ORDERDATE dtype:", df['ORDERDATE'].dtype)
print("Date range:", df['ORDERDATE'].min(), "to", df['ORDERDATE'].max())

# Re-check missing values after dropping -- updated missing value summary
missing = df.isnull().sum()
missing_pct = (df.isnull().mean()*100).round(2)
pd.concat([missing, missing_pct], axis=1).rename(columns={0:'missing_count',1:'missing_pct'})
```

```python
# Step 3: Drop high-missing columns
df_clean = df.drop(columns=['STATE', 'TERRITORY', 'POSTALCODE'])

# Check new shape and info
print("New shape:", df_clean.shape)
print("\n--- INFO ---")
df_clean.info()

print("\n--- Missing values (count & %) ---")
missing = df_clean.isnull().sum()
missing_pct = (df_clean.isnull().mean() * 100).round(2)
missing_df = pd.concat([missing, missing_pct], axis=1)
missing_df.columns = ['missing_count', 'missing_pct']
display(missing_df[missing_df['missing_count'] > 0].sort_values('missing_count', ascending=False))

print("\n--- Number of unique values per column ---")
display(df_clean.nunique().sort_values(ascending=False).to_frame(name='n_unique'))

print("\n--- Numeric summary (describe) ---")
display(df_clean.describe())
```

```python
# Data exploration/ Visualization:
# Histograms of numeric columns (SALES, QUANTITYORDERED, PRICEEACH, MSRP).
# Bar plots of categorical columns (PRODUCTLINE, STATUS, DEALSIZE, COUNTRY).

# Step 3: Univariate Analysis (Histograms + Bar Charts)

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Numeric columns  Histograms
numeric_cols = ['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']

df_clean[numeric_cols].hist(bins=20, figsize=(14, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Columns", fontsize=16)
plt.show()

# 2. Categorical columns  Bar charts (top categories if many)
categorical_cols = ['PRODUCTLINE', 'STATUS', 'DEALSIZE', 'COUNTRY']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    order = df_clean[col].value_counts().index  # sort by count
    sns.countplot(x=col, data=df_clean, order=order, ax=axes[i], palette='Set2')
    axes[i].set_title(f"Count of {col}")
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()


```

### after seeing output:
### QUANTITYORDERED: Mostly between 2050 items per order; very few extreme values beyond 60.
### PRICEEACH: Big spike at 100 (many products priced at 100); otherwise spread between ~3090.
### SALES: Right-skewed; most orders below 5000 USD but some very large ones (up to 14k).
### MSRP: Bimodal-ish (two peaks) around 75125, few very high MSRP products (above 175).
### PRODUCTLINE: Classic Cars dominates, then Vintage Cars. Trains are rare.
### STATUS: Vast majority Shipped; very few cancelled, on hold, disputed.
### DEALSIZE: Mostly Medium and Small orders, very few Large.
### COUNTRY: USA dominates, then Spain and France; long tail of many countries with small counts.




```python
# Bivariate analysis:
# Correlations between numeric variables.
# SALES vs QUANTITYORDERED, SALES vs PRODUCTLINE.
# SALES by YEAR_ID and by QUARTER.
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Quantity vs Sales
plt.figure(figsize=(8,6))
sns.scatterplot(x='QUANTITYORDERED', y='SALES', data=df_clean, alpha=0.5)
plt.title('Quantity Ordered vs Sales')
plt.show()

# Boxplot: Product Line vs Sales
plt.figure(figsize=(12,6))
sns.boxplot(x='PRODUCTLINE', y='SALES', data=df_clean)
plt.xticks(rotation=45)
plt.title('Sales Distribution by Product Line')
plt.show()

# Boxplot: Deal Size vs Sales
plt.figure(figsize=(8,6))
sns.boxplot(x='DEALSIZE', y='SALES', data=df_clean)
plt.title('Sales by Deal Size')
plt.show()

# Correlation Heatmap for numeric variables
plt.figure(figsize=(8,6))
numeric_cols = df_clean.select_dtypes(include=['float64','int64']).columns
corr = df_clean[numeric_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

### 1. QUANTITYORDERED increases, SALES also increase but sale depends on PRICEEACH, so some points with low quantity but price appear higher up.

### 2. sales Dist. by Product Line (box Plot) show median sale per order per product line, classic cars have highest median & widest spread, meaning some very deals, motorcycles are high but smaller spread. Trains and ships has lowest medians. thats mean product line strongly influences order size.

### 3. Sales by Deal size (box plot) -- large deal size orders have far higher sales than medium or small, these are few outliers but pattern is obvious

### 4. Correlation Heatmap shows :
#### QUANTITYORDERED moderately correlates with SALES (0.55).
#### PRICEEACH also correlates with SALES (0.66).
#### PRICEEACH and MSRP correlate (0.67) as expected.
#### QTR_ID & MONTH_ID are highly correlated (0.98)  redundant if used for modeling.
#### YEAR_ID and ORDERNUMBER correlate strongly (0.90)  later order numbers come in later years.

```python
# Time series:
# Trend of SALES over time (ORDERDATE).
df_clean['ORDERDATE'] = pd.to_datetime(df_clean['ORDERDATE'])

# Extracting useful time parts
df_clean['OrderYear'] = df_clean['ORDERDATE'].dt.year
df_clean['OrderMonth'] = df_clean['ORDERDATE'].dt.month
df_clean['OrderQuarter'] = df_clean['ORDERDATE'].dt.to_period('Q')

# 1. Sales trend by year
plt.figure(figsize=(8,5))
df_clean.groupby('OrderYear')['SALES'].sum().plot(marker='o')
plt.title("Total Sales by Year")
plt.ylabel("Total Sales")
plt.xlabel("Year")
plt.grid(True)
plt.show()

# 2. Monthly sales trend (all years combined)
plt.figure(figsize=(10,5))
df_clean.groupby('OrderMonth')['SALES'].sum().plot(marker='o')
plt.title("Total Sales by Month (all years combined)")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.grid(True)
plt.show()

# 3. Quarterly sales trend (all years combined)
plt.figure(figsize=(10,5))
df_clean.groupby('OrderQuarter')['SALES'].sum().plot(kind='bar', color='skyblue')
plt.title("Total Sales by Quarter")
plt.ylabel("Total Sales")
plt.xlabel("Quarter")
plt.show()

# 4. Sales over time (daily)
plt.figure(figsize=(14,6))
daily_sales = df_clean.groupby('ORDERDATE')['SALES'].sum()
daily_sales.plot()
plt.title("Daily Sales Over Time")
plt.ylabel("Sales")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

Total Sales by Year

2003  3.5 M, 2004  4.7 M, 2005  1.8 M

 Peak sales in 2004, sharp drop in 2005 (dataset ends in mid-2005).

Total Sales by Month (all years combined)

Months 1011 (OctNov) are highest; June is lowest.

 Clear seasonality; strong year-end demand.

Total Sales by Quarter

Q4 of each year dominates, especially 2004Q4.

Q1 and Q2 weaker.

 Classic seasonal sales cycle.

Daily Sales Over Time

Sales are spiky (large individual orders).

General upward trend from 2003 to 2004; more volatility after 2004.

Drop early 2005 due to fewer rows / truncated data.
